### Key patterns

**Length is the strongest quality signal.** All 6 degenerate outputs were >1000 chars. All 8 of the highest-scored (5) examples were <600 chars. The dataset's source model degrades reliably in long outputs.

**Quality is bipolar, not uniform.** 37% of examples score 4-5, 44% score 1-2, only 20% are middling. This is actually good for filtering — bipolar distributions separate cleanly.

**Format-specific tasks are systematically broken.** Both haiku requests (#20, #25) failed identically. Likely all poetry-form requests in this dataset are unreliable.

**Code/technical examples leak English.** When asked technical questions, the source model often defaulted to English output (#14) or partially-translated code terminology (#27).

### Recommended filters for Week 2

1. Drop outputs > 700 chars (catches degeneration — biggest impact)
2. Drop outputs with <30% Devanagari characters (catches English leakage)
3. Drop outputs where `output.strip() == input.strip()` (identity outputs)
4. Drop outputs with 3+ repetitions of any 5+ token n-gram (repetition loops)
5. Drop outputs starting with "मैं एक एआई" or "एक भाषा मॉडल के रूप में" (AI refusal patterns)
6. Drop examples flagged for harmful content (SQL injection, hacking, etc.)

**Estimated retention after filters:** 40-50% of 52K = ~21-26K usable examples.
