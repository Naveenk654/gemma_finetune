## Final Summary: Dataset Selection

**Approach:** Manually scored 30 + 15 + 10 random samples across three candidate datasets using a 1-5 quality rubric aligned to project goal (general-purpose Hindi/Hinglish instruction-following assistant).

### Dataset comparison

| Dataset | Samples | Avg Quality | Decision | Notes |
|---------|---------|-------------|----------|-------|
| iamshnoo/alpaca-cleaned-hindi | 30 | 2.93/5 | Use with aggressive filtering | 10 distinct failure modes; tail-degeneration in long outputs |
| rishiraj/samvaad-hi-v1 | 15 | ~4.4/5 (Hindi-only) | Use as primary | Multi-turn, Indian context; ~73% Hindi conversations |
| ai4bharat indic-instruct (wikihow_hi) | 10 | 5.0/5 | Use heavily | Zero failures; URL-encoded title bug (easy fix) |

### Failure modes identified across datasets (used to design Week 2 filters)
1. degenerate_long_output (Alpaca, common)
2. lazy_output, identity_output (Alpaca)
3. repetition_loop (Alpaca)
4. english_output (Alpaca, code/technical)
5. ai_refusal (Alpaca)
6. wrong_format (Alpaca, poetry tasks)
7. wrong_reasoning (Alpaca)
8. translation_artifact (Alpaca, samvaad partial)
9. broken_instruction (Alpaca)
10. url_encoded_text (wikihow titles — systematic, single-line fix)
11. english_only_conversations (samvaad)

### Final training mix (locked)
- 40% ai4bharat wikihow_hi (~12K examples)
- 40% samvaad-hi-v1 Hindi-filtered (~12K examples)
- 20% iamshnoo Alpaca filtered (~6K examples)
- **Target:** ~30K cleaned, deduplicated training pairs
